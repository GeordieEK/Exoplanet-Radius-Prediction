```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

# so that we can see all the columns
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
# set float number display format
pd.set_option('display.float_format', lambda x: '%.9f' % x)

# read csv into a dataframe
df = pd.read_csv('exoplanets_data.csv', sep = ',')
df.head()
```

## Drop rowid column

```python
# drop rowid column
df = df.drop('rowid', axis=1)
```

## Drop empty columns

```python
# drop empty columns
nunique = df.nunique()
cols_to_drop = nunique[nunique < 1].index
df = df.drop(cols_to_drop, axis=1)
```

```python
# drop rows with false positives
df = df.drop(df[[df['koi_disposition'] == 'FALSE POSITIVE']])
```

These tables are of the err1 or err2 values respectively. err1_pc and err2_pc are the percentage versions.

```python
# get all error columns
err1 = df.filter(regex='_err1', axis=1) # All columns that contain error 1 values
err2 = df.filter(regex='_err2', axis=1) # All columns that contain error 2 values
columns_with_error_range = err1.columns.str.removesuffix("_err1") # All columns that have data that could be inaccurate

err1_pc = err1.div(df[columns_with_error_range].values, axis='columns') * 100
err2_pc = err2.div(df[columns_with_error_range].values, axis='columns') * 100

total_err = pd.concat([err1, err2])
total_err.median().abs()
```
